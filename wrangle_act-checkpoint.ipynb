```python
# Dogs predection..


plt.figure(figsize=(10,6))
df_master['predection'].value_counts()[0:10].sort_values(ascending=True).plot(kind='barh' , color ='orange')

# declare the title of the plot and (x,y)
plt.xlabel('count',size=12)
plt.ylabel('predection', size=12)
plt.title('Dog\'s predection' , size=12)

x = df_master['predection'].value_counts()[0:10].sort_values(ascending=True).values
y= np.arange(0,10,1)

# to set the values on each bar
for x1, y1 in zip(x,y):
    plt.text(x1+1,y1,x1 , va='center')
```
